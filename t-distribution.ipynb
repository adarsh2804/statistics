```python
mu = 0
sigma = 1
N = 100000
popn = np.random.normal(loc=mu,scale=sigma,size = N)
```

```python
sn.distplot(popn,hist=False)
stats.shapiro(popn)
```

```python
def create_t_dist(popn,loops,n):
    t_seris = np.zeros(loops)
    for i in np.arange(loops):
        sampl = np.random.choice(popn, size=n)
        x_bar = np.mean(sampl)
        s = np.std(sampl)
        t = (x_bar - mu)/(s/np.sqrt(n))
        t_seris[i] = t
    return t_seris
```

```python
t_5 = create_t_dist(popn, 10000,5)
#t_10 = create_t_dist(popn, 10000,10)
t_20 = create_t_dist(popn, 10000,20)
#t_40 = create_t_dist(popn, 10000,40)
#t_150 = create_t_dist(popn, 10000,150)
```

```python
sn.distplot(popn,hist=False, label = "std normal");
sn.distplot(t_5,hist=False, label = "n=5");
#sn.distplot(t_10,hist=False, label = "n=10");
sn.distplot(t_20,hist=False, label = "n=20");
#sn.distplot(t_150,hist=False, label = "n=150");
plt.legend()
```
